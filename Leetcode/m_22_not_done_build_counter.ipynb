```python
# return n*(n-1)...*(n-t)
def p_n(n,t):
    if t <2:
        return n
    return n*p_n(n-1,t-1)
# return n!
def p_n2(n):
    if n<2:
        return n
    return n*p_n2(n-1)
# C n1 取 n2
def c_n(n1,n2):
    return int(p_n(n1,n2)/p_n2(n2))

```
