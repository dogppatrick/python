```python
import pandas as pd
import numpy as np
raw_data = pd.read_csv("push_gt50_cf.csv", encoding="utf-8")
n_raw =raw_data.shape[0]
raw_data
```

```python
dn = np.array(raw_data)
data_gt10 = dn[dn[:,2]>10,:]
```

```python
len(data_gt10)
s1_list = np.unique(data_gt10[:,1])
```

```python
c = ["id1", "id2", "f_rate", "ins12", "s1", "s2"]
df = pd.DataFrame(columns=c)
n = len(data_gt10)
# n = 10
# s1_tmp = s1_list[:10]
# print(s1_tmp)
# data_d = {}
for s1 in s1_list:
    s1_data = data_gt10[data_gt10[:,0]==s1, :]
    s1_fans = s1_data[s1_data[:,2].argsort()][-3:,:]
    for d in s1_fans:
        s = pd.Series(d, index=c)
        df = df.append(s, ignore_index=True)
```

```python
df.to_csv("push_gt50_r3.csv", index=False, encoding="utf-8" )
```
