```python
import pandas as pd
raw_data = pd.read_csv("push_data_g500.csv", encoding="utf-8")
# push >500
p_id = pd.read_csv("p_id_500.csv", encoding="utf-8")
n_raw =raw_data.shape[0]
```

```python
p_list = list(p_id["p_id"])
```

```python
# compare each ID
import time
tStart = time.time()#
data_out=[]
for idx in range(10):
    c = 0
    for i in range(n_raw):
        d = raw_data.iloc[i]
        art = d["uuid"]
        p_id = d["p_id"]
        if p_id == p_list[idx]:
            if c == 0: 
                data_out.append(set())
            else:
                data_out[idx].add(art)
            c = c + 1
        if c >100:
            break
print(time.time()-tStart)
# data_out
```

```python
for i in range(10):
    p1 = p_list[i]
    for j in range(10):
        p2 = p_list[j]
        if i !=j:
            ints = data_out[i].intersection(data_out[j])
            print(p1, "\t", p2, "\n ins:", len(ints), "total", len(data_out[j]) )
```
