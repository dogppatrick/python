```python
import pandas as pd
raw_data = pd.read_csv("push_data_too_big.csv", encoding="utf-8")
# push >100
p_id = pd.read_csv("p_id_500.csv", encoding="utf-8")
n_raw =raw_data.shape[0]
raw_data
```

```python
list(p_id["p_id"])
```

```python
n = n_raw
import time
tStart = time.time()#
target_id = list(p_id["p_id"])
c = ["uuid", "p_id"]
tg_data = pd.DataFrame(columns=c)
# n = 20000
for i in range(n):
    read_id = raw_data.iloc[i]["p_id"]
#     print(read_id)
    if read_id in target_id:
        save_data = [raw_data.iloc[i]["uuid"],raw_data.iloc[i]["p_id"]]
        s = pd.Series(save_data, index=c)
        tg_data = tg_data.append(s, ignore_index=True)
print(time.time()-tStart)
```

```python
tg_data.to_csv("push_data_g500.csv", index=False, encoding="utf-8" )
```
