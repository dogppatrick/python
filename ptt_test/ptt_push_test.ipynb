```python
import pandas as pd
raw_data = pd.read_csv("push_data_too_big.csv", encoding="utf-8")
push_data = pd.read_csv("push_count.csv", encoding="utf-8")
n_raw =raw_data.shape[0]
raw_data
```

```python
# 處理推文數量大於 120
# count_row = df.shape[0]  # gives number of row count
# count_col = df.shape[1]  # gives number of col count
push_count = {}
for i in range(n_raw):
    d = push_data.iloc[i,]
    p_id = d["p_id"]
    count = d["cnt"]
    if count > 120:
        push_count[p_id] = count
push_count
```
