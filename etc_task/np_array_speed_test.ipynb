```python
import pandas as pd
import numpy as np
import random as rd
# making testing data
c = ["c1", "c2", "c3", "c4"]
df = pd.DataFrame(columns=c)
for i in range(10000):
    data = [i, rd.randrange(1,5), rd.randrange(1,5), rd.randrange(1,50)]
    s = pd.Series(data, index=c)
    df = df.append(s, ignore_index=True)
```

```python
import time
arr = np.array(df)
df
```

```python
# time check
arr = np.array(df)
target = []
# q1 get c1 = specfic value
for i in range(10):
    target.append(rd.randrange(1, 10000))
print("find :", target)
# a1 by loop
print("============ a1 by loop ============")
a1_t1 = time.time()
for i in range(10000):
    if df.iloc[i]["c1"] in target:
#         print("got", df.iloc[i])
        print("got", i)
        print("time:", (time.time()-a1_t1))
print("all done time:", time.time()-a1_t1)
print("============ a2 by array ============")
a2_t1 = time.time()
for j in range(10):
    print("got", arr[arr[:,0]==target[j], :])
print("time:", (time.time()-a2_t1))
```
