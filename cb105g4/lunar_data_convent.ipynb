```python
import pandas as pd 
import numpy as np
df = pd.read_csv("lunar_c_utf8.csv", encoding="utf-8")
```

```python
# input data as array
arr_todo = df["宜"]
arr_ntodo = df["忌"]
inp_tag = ["嫁娶", "祭祀", "安葬", "開市"]
todo_new = []
ntodo_new = []
for i in range(len(arr_todo)):
    contex = arr_todo[i].replace("\"", "").split(",")
#     嫁娶 祭祀 安葬 開市
    todo_new.append([0, 0, 0, 0])
    for tag in contex:
        if tag in inp_tag:
            todo_new[i][inp_tag.index(tag)] = 1

for i in range(len(arr_ntodo)):
    contex = arr_ntodo[i].replace("\"", "").split(",")
#     嫁娶 祭祀 安葬 開市
    ntodo_new.append([0, 0, 0, 0])
    for tag in contex:
        if tag in inp_tag:
            ntodo_new[i][inp_tag.index(tag)] = 1
```

```python
# result to array
arr_todo = np.asarray(todo_new)
arr_ntodo = np.asarray(ntodo_new)
for i in range(len(inp_tag)):
    tag_g = "宜_" + inp_tag[i]
#     tag_b = "忌_" + inp_tag[i]
    df[tag_g] = arr_todo[:,i]
#     df[tag_b] = arr_ntodo[:,i]
for i in range(len(inp_tag)):
#     tag_g = "宜_" + inp_tag[i]
    tag_b = "忌_" + inp_tag[i]
#     df[tag_g] = arr_todo[:,i]
    df[tag_b] = arr_ntodo[:,i]
df
```
