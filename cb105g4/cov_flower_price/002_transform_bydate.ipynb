```python
# seting up 
import os, glob
import numpy as np
import pandas as pd
fn_list = glob.glob("../result/flat*.csv")
print(fn_list)
fn = fn_list[0]
```

```python
# read data
df = pd.read_csv(fn, encoding="utf-8")
df
```

```python
fn = fn.split("\\")[-1]
print(fn)
# df
# 總行數
nraw = df.shape[0]
# 個別欄位名稱
c = df.columns.tolist()
print(nraw)
for i in range(len(c)):
    print(i, c[i])
```

```python
# 先拉出 台北的資料測試
df_tp = df.groupby("市場").get_group("105 台北花市")
df_tp
```

```python
for colum in ['產品']:
    freq = df_tp[colum].value_counts()
    dist_value = df_tp[colum].drop_duplicates()
    print(colum,"總量",str(len(freq)))
#     print(freq)
#     print(dist_value)
    print("========")
p_list = dist_value
for colum in ['日期']:
    freq = df_tp[colum].value_counts()
    dist_value = df_tp[colum].drop_duplicates()
    print(colum,"總量",str(len(freq)))
#     print(freq)
#     print(dist_value)
    print("========")
date_list = dist_value
for colum in ['價別']:
    freq = df_tp[colum].value_counts()
    dist_value = df_tp[colum].drop_duplicates()
    print(colum,"總量",str(len(freq)))
#     print(freq)
#     print(dist_value)
    print("========")
ptype_list = dist_value
# print(p_list, date_list, ptype_list)
```

```python
arr_tp = np.array(df_tp)
arr_tp
```

```python
data_series = []
c = []
price_table = []
c.append("日期")
for flower in p_list:
    if flower.split(" ")[-1] == "":
        sub_flower_name=flower.split(" ")[-2]
    else:
        sub_flower_name=flower.split(" ")[-1]
    for p_type in ptype_list:
        c.append(p_type + "_" + sub_flower_name)
for date in date_list:
    price_series = []
    data_series=[]
    date_b = arr_tp[:,0]==date
    data_series.append(date)
    for flower in p_list:
        flower_b = arr_tp[:,2]==flower
        for p_type in ptype_list:
            flower_p_b = arr_tp[:,3]==p_type
            all_c = (date_b * flower_b * flower_p_b)
            if any(all_c):
                data_series.append(arr_tp[all_c,4])
#                 print(arr_tp[all_c,4][:])
            else:
                data_series.append("")
    price_table.append(data_series)
```

```python
t = np.asarray(price_table)
```

```python
# type(t)
df_tp_done = pd.DataFrame(t, columns=c)
df_tp_done
```

```python
df_tp_done.to_csv("test_wide_0424", index=False, encoding="utf-8" )
```
