```python
# seting up 
import os, glob
import numpy as np
import pandas as pd
fn_list = glob.glob("../result/flat*.csv")
print(fn_list)
fn = fn_list[0]
```

```python
# read data
df = pd.read_csv(fn, encoding="utf-8")
# df
```

```python
fn = fn.split("\\")[-1]
print(fn)
# df
# 總行數
nraw = df.shape[0]
# 個別欄位名稱
c = df.columns.tolist()
print(nraw)
for i in range(len(c)):
    print(i, c[i])
```

```python
# 先拉出 台北的資料測試
df_tp = df.groupby("市場").get_group("105 台北花市")
# df_tp
```

```python
df_tp['種類'] = df_tp[['產品', '價別']].apply(lambda x: '_'.join(x), axis=1)
df_tp
```

```python
df_w = df_tp.pivot_table(index="日期", columns="種類", values="價格", dropna=True)
# df_w["日期"] = df_w.index.values
# dfp.set_index()
df_w
```

```python
df_w.to_csv("test_pivot0425.csv", index=True)
```

```python
df = pd.read_csv("test_pivot0425.csv", encoding="utf-8")
df
```

```python
# 先將第一筆資料補上>>內插
data_0= df.iloc[0]
null_list = data_0.isnull()
i = 1
while any(null_list):
    data_up = df.iloc[i]
    data_0[null_list] = data_up[null_list]
    null_list = data_0.isnull()
    i = i +1
df.iloc[0]=data_0
df = df.interpolate()
df
```

```python
df.to_csv("test_pivot0425_interpolate.csv")
```
