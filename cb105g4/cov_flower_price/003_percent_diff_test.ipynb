```python
# 讀取檔案 先拆開日期 
df = pd.read_csv(fn, encoding="utf-8", index_col=False)
date = df["日期"]
df = df.drop(columns="日期")
# 做diff
df = df.pct_change()
df["日期"] = date
df = df.drop(index = 0)
df
```

```python
# t = df["FB001 火鶴花 混色_上價"]
df.describe()
```

```python
fn_save = fn.split("inter")[0] + "diff_" + fn.split("_")[-2] + "_" + fn.split("all")[-1]
print(fn_save)
df.to_csv(fn_save, encoding="utf-8", index=False)
```

```python
from scipy import spatial
tags = df.drop(columns="日期").columns.to_list()
cos_data=[]
for tag1 in tags:
    for tag2 in tags:
        vector1 = df[tag1]
        vector2 = df[tag2]
        result = [tag1, tag2, 1 - spatial.distance.cosine(vector1,vector2)]
        cos_data.append(result)
#         print("cosin:", result)
cos_data
```

```python
import matplotlib.pyplot as plt
df = pd.DataFrame(cos_data, columns=["t1","t2","value"])
df_p = df.pivot_table(index="t1", columns="t2", values="value")
df_p
```

```python
a = np.array(df_p)
s_n = 57
sample = a[0:s_n:3,0:s_n:3]
c = df_p.index.to_list()
c_s = c[0:s_n:3]
df_s = pd.DataFrame(sample, columns=c_s, index=c_s)
data = np.array(df_s)

plt.pcolor(df_s)
# plt.yticks(c_s)
# plt.xticks(c_s)
print(c_s)
```
