```python
market_list = ["台北", "台中", "台南", "高雄", "彰化"]
flower_list = ["Anthurium", "Chrysanthemum", "Eustoma", "OrientalLily", "Rose"]
data_all = []
d_s = date(2019, 6, 14)
for market in market_list:
    for flower in flower_list:
        for i in range(2):
            data = [market,flower]
            d2 = (d_s + timedelta(days=i)).strftime("%Y-%m-%d")
            data.append(d2)
            try:
                p = PPModel(market, d2, flower, 3, 1)
                data.append(p.result())
#                 print(data)
            except:
                continue
        
            data_all.append(data)
print("done")
```

```python
df_add =  pd.DataFrame(data_all, columns=["market", "flower", "date", "price_predict"])
df_old =  pd.read_csv("price_predict0531.csv", encoding="utf-8", names=["market", "flower", "date", "price_predict"])
df_all = df_old.append(df_add)
```
