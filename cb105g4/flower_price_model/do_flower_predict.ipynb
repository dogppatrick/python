```python
from price_predict_model import PPModel
from datetime import date, timedelta
import pandas as pd
import numpy as np
d_s = date(2019, 6, 4)
market_list = ["台北", "台中", "台南", "高雄", "彰化"]
flower_list = ["Anthurium", "Chrysanthemum", "Eustoma", "OrientalLily", "Rose"]
data_all = []
for market in market_list:
    for flower in flower_list:
        for i in range(10):
            data = [market,flower]
            d2 = (d_s + timedelta(days=i)).strftime("%Y-%m-%d")
            data.append(d2)
            try:
                p = PPModel(market, d2, flower, 3, 1)
                data.append(p.result())
                print(data)
            except:
                continue
        
            data_all.append(data)
        
```

```python
df = pd.DataFrame(data_all, columns=["market", "flower", "date", "price_predict"])
df.to_csv("price_predict0531.csv", encoding="utf-8", index=False, header=False)
```
