```python
# 已開 * 有開統一發票 * 8 殯葬 9 餐廳 10 花卉銷售 # 11 花卉栽種
c_with_uni_date=[]
for i in range(total_d):
    date = cdate[i]
    b_op = opdata[i]*1 # 該時間點已開張的條件
    b_uni = (d_arr[:, 3]=="Y")*1
    result = []
    result.append(date)
    for c in range(8,11+1):
        b_c = (d_arr[:, c]==1)*1
        b = (b_op*b_uni*b_c).astype(bool)
        result.append(len(d_arr[b, :]))
    c_with_uni_date.append(result)
    
```
