```python
c = 0
data_w = ""
find_list = ["花店", "花卉", "鮮花", "盆栽", "園藝"]

with open("D:/Python/cb105g4/data_source/business_live.csv", "r", encoding="utf-8") as f:
    data = f.readline()
#     add title
    data_w = data_w + data
    while data != "":
        data = f.readline()
        if any(item in data for item in find_list):
            data_w = data_w + data
            c = c + 1

        if ((c+1) % 1000) == 0:
            with open("D:/Python/cb105g4/fc_live.csv", "a", encoding="utf-8") as fr:
                fr.write(data_w)
                data_w = ""

                print("data_saved")
    with open("D:/Python/cb105g4/fc_live.csv", "a", encoding="utf-8") as fr:
        fr.write(data_w)
```
