```python
def weather_join_flower(w_s, flower):
    print(w_s, flower)
    fn_w = glob("../result/merge_weather/"+ w_s +"*.csv")[0]
    fn_f = glob("../result/merge_flower_allmarket_price/merge_"+ flower + "*.csv")[0]
    df_w = pd.read_csv(fn_w, encoding="utf-8")
    df_f = pd.read_csv(fn_f, encoding="utf-8")
    df_j = df_w.join(df_f.set_index("date"), on="Date")
    fn_save = "../result/join_weather_flower/" + flower + "_" + w_s + ".csv"
    df_j.to_csv(fn_save, index=False, encoding="utf-8")
    print("saved")
```

```python
flowers = list(weather_flower.keys())
for flower in flowers:
    w_s = weather_flower[flower][0]
    weather_join_flower(w_s, flower)
```
