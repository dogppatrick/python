```python
import pandas as pd
import numpy as np
fn = "fc_live.csv"
data = pd.read_csv(fn, encoding="utf-8")
# data
# 新增 arr_d type np.array 用於計算
arr_d = np.array(data)
# c = 個別欄位名稱
c = data.columns.tolist()
# for i in range(len(c)):
#     print(c[i],i, end="\t")
#     print(arr_d[0:10,i])
data_raws = data.shape[0]
data
```

```python
import time
# 行業代號
b_c=np.array(list(range(8,14+1,2)))
# 名稱
b_n=np.array(list(range(9,15+1,2)))
bs_code_new = []
bn_code_new = []
t1 = time.time()
for i in range(data_raws):
    # l = 行業代號 去除nan 轉為int
    l = arr_d[i,b_c]
    l = l[~pd.isna(l)]
    l = l.astype(int)
    bs_code_new.append(l)
    # l = 名稱 去除nan
    l = arr_d[i,b_n]
    l = l[~pd.isna(l)]
    bn_code_new.append(l)
print(time.time()-t1)
# print(bs_code_new,bn_code_new)
data["行業代碼_new"]=bs_code_new
data["名稱_new"]=bn_code_new
data = data.drop(['名稱', '名稱1', '名稱2', '名稱3', '行業代號', 
                  '行業代號1', '行業代號2', '行業代號3', '總機構統一編號'], axis=1)
```
